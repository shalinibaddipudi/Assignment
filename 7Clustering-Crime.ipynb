```python
from sklearn.preprocessing import MinMaxScaler
transformation=MinMaxScaler()
Crime=pd.DataFrame(transformation.fit_transform(crime.iloc[:,1:]))
Crime
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
l=linkage(Crime,method="average",metric="euclidean")
plt.figure(figsize=(15,5))
plt.title("Hierarchinal clustering Dendrogram")
plt.xlabel("Index")
plt.ylabel("Distance")
sch.dendrogram(l)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
H_complete=AgglomerativeClustering(n_clusters=4, linkage="average",affinity="euclidean").fit(Crime)
cluster_labels=pd.Series(H_complete.labels_)
cluster_labels
crime['clust']=cluster_labels
crime
```

### K-means clustering

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(Crime)
    WCSS.append(clf.inertia_) 
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show() 
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(Crime)  
```

```python
y_kmeans
clf.labels_ 
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md=pd.Series(y_kmeans) 
crime['clust']=md  
crime
```

```python
crime.iloc[:,1:].groupby(crime.clust).mean()
```

```python
crime.plot(x="Murder",y ="Assault",c=clf.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors=NearestNeighbors(n_neighbors=8)
nbrs=neighbors.fit(Crime)
distances, indices=nbrs.kneighbors(Crime)
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```

```python
dbscan=DBSCAN(eps=0.315, min_samples=5)
dbscan.fit(Crime)
```

```python
dbscan.labels_
```

```python
d1=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
d1
pd.set_option("display.max_rows",None)
```
