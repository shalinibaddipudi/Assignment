## Variation Infliation Factor:VIF

```python
rsq_age=smf.ols('Age~KM+HP+CC+Doors+Gears+Weight+QT',data=toyota).fit().rsquared
rsq_km=smf.ols('KM~Age+HP+CC+Doors+Gears+Weight+QT',data=toyota).fit().rsquared
rsq_hp=smf.ols('HP~KM+Age+CC+Doors+Gears+Weight+QT',data=toyota).fit().rsquared
rsq_cc=smf.ols('CC~KM+HP+Age+Doors+Gears+Weight+QT',data=toyota).fit().rsquared
rsq_doors=smf.ols('Doors~KM+HP+CC+Age+Gears+Weight+QT',data=toyota).fit().rsquared
rsq_gears=smf.ols('Gears~KM+HP+CC+Doors+Age+Weight+QT',data=toyota).fit().rsquared
rsq_weight=smf.ols('Weight~KM+HP+CC+Doors+Gears+Age+QT',data=toyota).fit().rsquared
rsq_qt=smf.ols('QT~KM+HP+CC+Doors+Gears+Weight+Age',data=toyota).fit().rsquared
```

```python
vif_age=1/(1-rsq_age)
print("VIF of age:",vif_age)
vif_km=1/(1-rsq_km)
print("VIF of Km:",vif_km)
vif_hp=1/(1-rsq_hp)
print("VIF of Hp:",vif_hp)
vif_cc=1/(1-rsq_cc)
print("VIF of Cc:",vif_cc)
vif_doors=1/(1-rsq_doors)
print("VIF of Doors:",vif_doors)
vif_gears=1/(1-rsq_gears)
print("VIF of Gears:",vif_gears)
vif_weight=1/(1-rsq_weight)
print("VIF of Weight:",vif_weight)
vif_qt=1/(1-rsq_qt)
print("VIF of Qt:",vif_qt)
```

### Analysis of influential observations:Detection and Deletion

```python
model1_influence = model1.get_influence()
(c, _) = model1_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(20, 6))
plt.stem(np.arange(len(toyota)), np.round(c, 5))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(c),np.max(c)
```
